```python
import torch
import torch.nn as nn
```

```python
class Perceptron(nn.Module):
    """A perceptron is one linear layer"""
    def __init__(self, input_dim):
        super(Perceptron, self).__init__()
        self.fc1 == nn.Linear(input_dim, 1)
        
        def forward(self, x_in):
            """forward pass, backprop calculate automagically"""
            return torch.sigmoid(Self.fc1(x_in)).squeeze()
        
```

# Activation functions

# Sigmoid
**Usage**: mainly as ouput on final layer for probabilities. Usage in other layers can lead to vanishing or exploding grdients because of the distribution of the ouputs (near zero or overflowing floating point value)

```python
import matplotlib.pyplot as plt
```

```python
x = torch.arange(-5., 5., 0.1)
y = torch.sigmoid(x)
plt.plot(x.numpy(), y.numpy())
plt.show()
```

# Tanh
**Usage**: postive and negative numbers in output between -1 and 1

```python
x = torch.arange(-5., 5., 0.1)
y = torch.tanh(x)
plt.plot(x.numpy(), y.numpy())
plt.show()
```
